## Decision Tree Classifier

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['target'])
```

```python
X = pd.DataFrame(iris['data'], columns=['sepal length','sepal width','petal length','petal width'])
```

```python
y = pd.DataFrame(iris['target'],columns=['Type'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state=10)
```

```python
## applu dec tree classifier 
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
X_train
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(classification_report(y_pred, y_test))
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Decision Tree PrePuning

```python
params = { 
    'criterion' :[ "gini", "entropy", "log_loss"],
    'splitter' : ["best","random"],
    'max_depth': [1,2,3,4,5],
    'max_features' : ["auto","sqrt", 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
treegsv = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv = 5, verbose= 3, scoring='accuracy')
```

```python
treegsv.fit(X_train,y_train)
```

```python
treegsv.best_params_
```

```python
treegsv.best_score_
```

```python
y_pred2 = treegsv.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_pred2, y_test))
```
