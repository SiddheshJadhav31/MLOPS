```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('business.csv')
df.head()
```

```python
df.drop(columns=['stop'],inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
#First feature engineering process: Convert Date to date month and year
df['Date']= df['date'].str.split('-').str[0]
df['Month'] = df['date'].str.split('-').str[1]
df['Year'] = df['date'].str.split('-').str[2]
df.drop(columns=['date'],inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
df.head()
```

```python
df['arr_hr'] = df['arr_time'].str.split(':').str[0]
df['arr_min'] = df['arr_time'].str.split(':').str[1]
df['dept_hr'] = df['dep_time'].str.split(':').str[0]
df['dept_min'] = df['dep_time'].str.split(':').str[1]
df.head()
```

```python
df.drop(columns=['arr_time','dep_time'],inplace=True)
df.head()
```

```python
df['arr_hr'] = df['arr_hr'].astype(int)
df['arr_min'] = df['arr_min'].astype(int)
df['dept_hr'] = df['dept_hr'].astype(int)
df['dept_min'] = df['dept_min'].astype(int)
df.info()
```
